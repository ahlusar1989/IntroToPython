```python
#Let's Focus on Obama and Romney

fec_mitt_barack = fec[fec.cand_nm.isin(['Obama, Barack', 'Romney, Mitt'])]
```

```python
emp_mapping = {'INFORMATION REQUESTED PER BEST EFFORTS' : 'NOT PROVIDED',
               'INFORMATION REQUESTED' : 'NOT PROVIDED',
               'SELF' : 'SELF EMPLOYED',
               'SELF EMPLOYED' : 'SELF EMPLOYED',
              }
```

```python
f = lambda x: emp_mapping.get(x,x)
fec.contbr_employer = fec.contbr_employer.map(f)
```

```python
by_occupation = fec.pivot_table('contb_receipt_amt',
                                index = 'contbr_occupation',
                                columns = 'party', aggfunc = 'sum')
```

```python
over_2mm = by_occupation[by_occupation.sum(1) > 2000000]
```

```python
over_2mm
```

```python
over_2mm.plot(kind = 'bar', stacked=True)
```

```python
def get_top_amounts(group, key, n = 5):
    totals = group.groupby(key)['contb_receipt_amt'].sum()
    return totals.order(ascending = False)[-n:]
```

```python
grouped = fec_mitt_barack.groupby('cand_nm')
```

```python
grouped.apply(get_top_amounts, 'contbr_occupation', n = 10)
```

```python
grouped.apply(get_top_amounts, 'contbr_employer', n = 10)
```

```python
bins = np.array([0,1,10,100,1000,10000,100000,1000000,10000000])
```

```python
labels = pd.cut(fec_mitt_barack.contb_receipt_amt, bins)
```

```python
labels
```

```python
grouped = fec_mitt_barack.groupby(['cand_nm', labels])
```

```python
grouped.size().unstack(0)
```

```python
#Sum the Contribution amounts and normalize within buckets to visualize percentage of total donations of each size by 
#candidate

bucket_sums = grouped.contb_receipt_amt.sum().unstack(0) 
```

```python
bucket_sums
```

```python
normed_sums = bucket_sums.div(bucket_sums.sum(axis=1), axis = 0)
```
